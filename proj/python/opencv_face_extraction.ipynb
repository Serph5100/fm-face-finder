#### ==== **Face Detection and Cropping with OpenCV** ====

```python
import numpy as np
import cv2
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt
import util
%matplotlib inline
```

```python
query = 'roberto baggio juventus 1994'
image_number_in_query = 4
```

```python
img_loc = f'dataset/{query}/Image_{image_number_in_query}.jpg'
```

```python
sample_image = io.imread(img_loc)
plt.imshow(sample_image)
```

```python
test_image_gray = cv2.cvtColor(sample_image, cv2.COLOR_BGR2GRAY)
plt.imshow(test_image_gray, cmap='gray')
```

```python
haar_cascade_face = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
face_rectangles = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor=1.2, minNeighbors=5)
```

```python
print(f'Number of faces detected : {len(face_rectangles)}')
```

```python
test_image_RGB = cv2.imread(img_loc)
```

```python
# --- put rectangles in face image , plus the frame adjustments ---
for (x, y, w, h) in face_rectangles:
    x, y, w, h = int(x), int(y), int(w), int(h)
    cv2.rectangle(test_image_RGB, 
                  (x, y), 
                  (x + w, y + h), 
                  (255, 0, 0), 
                  2)
```

```python
plt.imshow(util.convertToRGB(test_image_RGB))
```
