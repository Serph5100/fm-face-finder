#### ==== **Cropping Faces with Autocrop library** ====

```python
from autocrop import Cropper
from pathlib import Path
import PIL
import matplotlib.pyplot as plt
```

```python
#Initialize Cropper
cropper = Cropper(
    height=310,
    width=260,
    face_percent=70,
)
```

```python
query = 'abedi pele torino 1994'
image_number_in_query = 3
```

```python
image_loc = f'dataset/{query}/Image_{image_number_in_query}.jpg'
```

```python
original_image = PIL.Image.open(image_loc)
```

```python
#see the size of the image
plt.imshow(plt.imread(image_loc))
```

```python
cropped_array = cropper.crop(image_loc)

#True if face detected, false if not
if cropped_array is not None:
    print("Face detected")
    cropped_image = PIL.Image.fromarray(cropped_array)
    cropped_image.show()
else:
    print("No face detected")
```
